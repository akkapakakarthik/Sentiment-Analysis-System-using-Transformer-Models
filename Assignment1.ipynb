```python
from scipy.io import loadmat
mnist = loadmat("/home/awez_mehtab/WiDS/mnist-original.mat")
mnist_data = mnist["data"].T
mnist_label = mnist["label"][0]
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

# Flattening and normalizing the pixel values, then we visualize a few examples from each digit class
for digit in range(10):
    #ing Filtering indices for the current digit
    digit_indices = (mnist_label == digit)
    
    # Creating a subplot for the current digit
    plt.figure(figsize=(15, 4))
    plt.suptitle(f'Examples from Digit {digit}', fontsize=16)

    # Iterating over the first 5 images for the current digit
    for i in range(5):
        # Extracting the flattened and normalized image and label
        flattened_image = mnist_data[digit_indices][i].flatten()
        normalized_image = flattened_image / 255.0  # Normalizing pixel values to the range [0, 1]
        label = mnist_label[digit_indices][i]

        # Displaying the normalized image
        plt.subplot(1, 5, i + 1)  # Setting up subplot grid with 1 row and 5 columns
        plt.imshow(normalized_image.reshape(28, 28), cmap='gray')  # Displaying the reshaped normalized image
        plt.title(f"Example {i+1}", fontsize=12)  # Showing example number as the title
        plt.axis('off')

    plt.show()

# Calculating and displaying the distribution of labels
plt.figure(figsize=(10, 5))  # Setting the size of the figure

# Calculating the histogram of labels
plt.hist(mnist_label, bins=range(11), align='left', rwidth=0.8, color='blue', edgecolor='black')

# Setting labels and title
plt.xlabel('Digit')
plt.ylabel('Count')
plt.title('Distribution of Labels (Digits 0-9)')

# Displaying the plot
plt.show()
```

## Task 2

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming mnist_data contains the image data and mnist_label contains the labels

# Generating 10 random indices
random_indices = np.random.choice(len(mnist_label), size=10, replace=False)

# Creating a subplot for the random sample
plt.figure(figsize=(15, 4))
plt.suptitle('Random Sample of 10 Images with True Labels', fontsize=16)

# Iterating over the 10 random indices
for i, idx in enumerate(random_indices):
    # Extracting the flattened and normalized image and label
    flattened_image = mnist_data[idx].flatten()
    normalized_image = flattened_image / 255.0  # Normalizing pixel values to the range [0, 1]
    label = mnist_label[idx]

    # Displaying the normalized image
    plt.subplot(2, 5, i + 1)  # Setting up subplot grid with 2 rows and 5 columns
    plt.imshow(normalized_image.reshape(28, 28), cmap='gray')  # Displaying the reshaped normalized image
    plt.title(f"True Label: {label}", fontsize=12)  # Showing the true label as the title
    plt.axis('off')  # Turning off axis ticks and labels

plt.show()

```

```python
import matplotlib.pyplot as plt

# Assuming mnist_label contains the labels

# Calculating the histogram of labels
label_counts = [np.sum(mnist_label == digit) for digit in range(10)]

# Creating a bar chart
plt.figure(figsize=(10, 5))
plt.bar(range(10), label_counts, color='blue', edgecolor='black')

# Setting labels and title
plt.xlabel('Digit')
plt.ylabel('Count')
plt.title('Distribution of Digit Classes in the Dataset')

# Setting x-axis ticks to represent digits 0 through 9
plt.xticks(range(10), [str(digit) for digit in range(10)])

# Displaying the plot
plt.show()
```
